```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("Olympics-Data.csv")
```

```python
df1 = data
```

```python
df1 = df1[['Country','Medal']]
```

```python
df1
```

```python
df1 = df1[df1['Medal'] == 'Gold']
```

```python
df1
```

```python
df1_gp = df1.groupby("Country")['Medal'].count()
```

```python
df1_gp
```

```python
df1_gp = df1_gp.sort_values(ascending = False)[:10]
```

```python
df1_gp
```

```python
list(df1_gp.index)
```

```python
list(df1_gp.values)
```

```python

fig_top10 , ax_top10 = plt.subplots(1,1,figsize=(10,6))
ax_top10.set_xticks(range(len(list(df1_gp.index))))
ax_top10.set_xticklabels(list(df1_gp.index))

ax_top10.set_yticks(range(0, 2501, 500))
ax_top10.tick_params(axis='x', rotation=45)
ax_top10.set_title('Top 10 Countries with the Most Gold Medals',fontsize = 12 , color = "#22333b")
ax_top10.set_ylabel('Number of Gold Medals',fontsize = 10,color = "#22333b")

ax_top10.spines['left'].set_visible(False)
ax_top10.spines['right'].set_visible(False)
ax_top10.spines['top'].set_visible(False)

ax_top10.set_facecolor("#EAF4f4")
fig_top10.set_facecolor("#EAF4f4")
ax_top10.tick_params(left=False , bottom = False,length = 0 ,labelcolor = "#22333b",color="#22333b")
plt.bar(range(len(list(df1_gp.index))), list(df1_gp.values), color = '#005f73')
plt.xticks(rotation=45)
plt.show()
```

```python
df3 = data
```

```python
df3 = df3[["Sport","Country","Medal"]]
```

```python
df3 = df3[df3["Country"] == "SVK"]
```

```python
df3
```

```python
df3 = df3[["Sport","Medal"]]
```

```python
df3
```

```python
df3_gp = df3.groupby(["Sport","Medal"])["Medal"].count()
```

```python
df3_gp
```

```python
df3_gp.index
```

```python
df3_gp.values
```

```python
df3_gp_total_medal = df3.groupby(["Sport"])["Medal"].count()
```

```python
df3_gp_total_medal
```

```python
df3_gp_total_medal.values
```

```python
labels_outer = list(df3_gp_total_medal.index)
```

```python
labels_outer
```

```python
fig_SVK, ax_SVK = plt.subplots(1,1,figsize = (10,8))

ax_SVK.set_facecolor("#EAF4f4")
fig_SVK.set_facecolor("#EAF4f4")

outer_colors = ['#d8f3dc', '#b7e4c7', '#95d5b2','#74c69d','#52b788','#40916c']
inner_colors_dict = {'Gold': '#E9C46A', 'Silver': '#E0E1DD', 'Bronze': '#CD7F32'}
inner_colors = [inner_colors_dict[medal] for sport, medal in df3_gp.index]
#print(inner_colors)
inner_labels = [f"{medal}" for sport, medal in df3_gp.index]
#print(inner_labels)

ax_SVK.pie(df3_gp_total_medal.values,labels = labels_outer, radius=1, colors=outer_colors,
       wedgeprops=dict(width=0.3, edgecolor='w'))

ax_SVK.pie(df3_gp.values,labels = inner_labels,labeldistance = 0.75, radius=0.7, colors=inner_colors,
       wedgeprops=dict(width=0.3, edgecolor='w'))
ax_SVK.set_title('Count of Medals in each Sport',fontsize = 12 , color = "#22333b")
ax_SVK.set(aspect="equal")
plt.show()
```
